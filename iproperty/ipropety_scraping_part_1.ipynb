```python
import cloudscraper
import time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
condo_name = []
condo_listing = []

bungalow_name = []
bungalow_listing = []

semi_detached_name = []
semi_detached_listing = []

terrace_name = []
terrace_listing = []
```

```python
property_type = {
    "condo" : "property_type=N",
    "bungalow" : "property_type=B",
    "semi-detached" : "property_type=S",
    "terrace" : "property_type=T"
}


def BSPrep(URL):
    exitcode = 1
    while exitcode == 1:
        try:
            trial = 0
            while trial < 10:
                scraper = cloudscraper.create_scraper()
                print('Loading '+URL)
                s = scraper.get(URL)
                soup = BeautifulSoup(s.content, 'html.parser')
                if "captcha" in soup.text:
                    trial += 1
                    print('Retrying '+' ('+str(trial)+'/10) ...')
                    time.sleep(0.1)
                    continue
                elif "No Results" in soup.text:
                    print('Invalid URL, skipping '+URL)
                    trial = 99
                else:
                    trial = 99
            if trial == 10:
                print('Trial exceeded, skipping '+URL)
            exitcode = 0
            return soup
        except:
                print('Connection reset, retrying in 1 mins...', flush=True)
                time.sleep(60)
                

## Get the listing link
def find_name_url(soup, array_name, array_listing):
    for a in soup.find_all('a', {"class": "nav-link"}):
        if "#contact-agent" not in a["href"]:
            array_name.append(a["title"])
            array_listing.append(a["href"])
            

## Get the data and push to defined list
def request_and_pust(property_input, array_name_input, array_listing_input):
    property_type = property_input
    property_name = array_name_input
    property_listing = array_listing_input
    
    for i in range(1,100):
        print(i, "==")
        soup = BSPrep(f"https://www.propertyguru.com.my/property-for-sale/{i}?freetext=Penang&{property_type}&region_code=MY07")
        find_name_url(soup, property_name, property_listing)
```

```python
## Condominium
request_and_pust(property_type["condo"], condo_name, condo_listing)
```

```python
## Bungalow
request_and_pust(property_type["bungalow"], bungalow_name, bungalow_listing)
```

```python
## Semi Detached
request_and_pust(property_type["semi-detached"], semi_detached_name, semi_detached_listing)
```

```python
## Terrace
request_and_pust(property_type["terrace"], terrace_name, terrace_listing)
```

```python
condo_df = pd.DataFrame(data={"property_name" : condo_name, "property_listing": condo_listing})

condo_df["property_type"] = "condominium"

condo_df.head()
```

```python
bungalow_df = pd.DataFrame(data={"property_name" : bungalow_name, "property_listing": bungalow_listing})

bungalow_df["property_type"] = "bungalow"

bungalow_df.head()
```

```python
semi_detached_df = pd.DataFrame(data={"property_name" : semi_detached_name, "property_listing": semi_detached_listing})

semi_detached_df["property_type"] = "semi_detached"

semi_detached_df.head()
```

```python
semi_detached_df.to_csv("test.csv", index=False)
```

```python
terrace_df = pd.DataFrame(data={"property_name" : terrace_name, "property_listing": terrace_listing})

terrace_df["property_type"] = "terrace"

terrace_df.head()
```

```python
property_df = pd.concat([condo_df, bungalow_df, semi_detached_df, terrace_df])
```
